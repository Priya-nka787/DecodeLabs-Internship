```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/dataset.csv')
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)
print("Dataset ready! Shape:", df.shape)
```

```python
df.describe()
```

```python
print("Survival Count:")
print(df['Survived'].value_counts())
print("\nSurvival Percentage:")
print(df['Survived'].value_counts(normalize=True) * 100)
```

```python
print("Survival by Gender:")
print(df.groupby('Sex')['Survived'].mean() * 100)
```

```python
print("Survival by Passenger Class:")
print(df.groupby('Pclass')['Survived'].mean() * 100)
```

```python
print("Age Statistics:")
print("Youngest passenger:", df['Age'].min())
print("Oldest passenger:", df['Age'].max())
print("Average age:", round(df['Age'].mean(), 2))
```

```python
print("Correlation with Survival:")
print(df.corr()['Survived'].sort_values(ascending=False))
```

## Key Findings
- Only 38% of passengers survived
- Women had 74% survival rate vs men 18%
- 1st class passengers survived more (63%) than 3rd class (24%)
- Youngest passenger was 0.42 years, oldest was 80
- Average age was 29.36 years

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/dataset.csv')
df.head()
```

```python
print("Missing values before cleaning")
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
print("Age missing values after fix:",df['Age'].isnull().sum())
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
print("Embarked missing values after fix:", df['Embarked'].isnull().sum())
```

```python
df.drop(columns= ['Cabin'],inplace = True )
print("Cabin column dropped")
```

```python
print("Duplicated before cleaning:", df.duplicated().sum())
```

```python
print("Dataset shape after cleaning:", df.shape)
```

## Summary
- Filled 177 missing Age values with median (28)
- Filled 2 missing Embarked values with mode (S)
- Dropped Cabin column (77% data missing)
- No duplicate rows found
- Final dataset shape: 891 rows × 11 columns
